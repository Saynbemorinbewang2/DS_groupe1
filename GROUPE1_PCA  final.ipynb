# L'analyse en composantes principales (PCA)
### L'analyse en composantes principales (PCA) est une technique statistique puissante utilisée pour simplifier et analyser des ensembles de données complexes. Son objectif est de réduire la dimensionnalité des données tout en conservant l'information importante.
# Applications du PCA dans la société
## 1. Reconnaissance faciale: 
### Le PCA est utilisé pour réduire la dimensionnalité des images faciales tout en conservant les caractéristiques importantes pour la reconnaissance.
## 2. Recommandation de produits: 
### Le PCA peut aider à identifier les caractéristiques les plus importantes pour les clients et à recommander des produits en fonction de leurs préférences.
## 3. Détection de fraude: 
### Le PCA peut être utilisé pour détecter des transactions frauduleuses en analysant les habitudes de dépenses des clients.
## 4. Analyse de texte: 
### Le PCA peut être utilisé pour réduire la dimensionnalité des documents textuels et faciliter l'analyse de contenu.
## 5. Compression d'images: 
### Le PCA peut être utilisé pour compresser des images en réduisant le nombre de dimensions nécessaires pour les représenter.
# Avantages du PCA:
## •	Réduction de la dimensionnalité: 
### Le PCA permet de simplifier les données en réduisant le nombre de dimensions tout en conservant l'information importante.
## •	Amélioration de la performance des modèles: 
### Le PCA peut améliorer la performance des modèles d'apprentissage automatique en réduisant la complexité des données.
## •	Détection de structures cachées: 
### Le PCA peut aider à identifier des structures et des patterns cachés dans les données.
## •	Interprétation des données: 
### Le PCA peut faciliter l'interprétation des données en les projetant sur un espace de dimensionnalité réduite.
# Inconvénients du PCA:
## •	Perte d'information: 
### La réduction de la dimensionnalité peut entraîner une perte d'information importante.
## •	Sensibilité aux outliers: 
### Le PCA est sensible aux valeurs aberrantes qui peuvent influencer les résultats de l'analyse.
## •	Interprétation difficile: 
### L'interprétation des composantes principales peut être difficile et sujette à caution.
# Concepts mathématiques du PCA:
### PCA se base sur les conceps suivant:
## •	Vecteurs propres et valeurs propres: 
### Le PCA calcule les vecteurs propres et les valeurs propres de la matrice de covariance des données.
## •	Transformation orthogonale: 
### Le PCA transforme les données dans un nouvel espace de dimensionnalité réduite défini par les vecteurs propres.
## •	Composantes principales: 
### Les composantes principales sont les nouvelles variables créées par la transformation PCA.
## •	Variance expliquée: 
### La variance expliquée par chaque composante principale est une mesure de son importance.
# Illustration des concepts mathématiques du PCA
## 1. Vecteurs propres et valeurs propres:
### Considérons la matrice de covariance suivante:
C = [[1 0.5]
     [0.5 1]]
### Les vecteurs propres de C sont les vecteurs v qui satisfont l'équation suivante:
Cv = λv
### où λ est un scalaire appelé valeur propre.
### Pour la matrice C, les vecteurs propres sont v1 = [1/√2, 1/√2] et v2 = [-1/√2, 1/√2] et les valeurs propres sont λ1 = 2 et λ2 = 0.
## 2. Transformation orthogonale:
### La transformation PCA consiste à projeter les données sur les vecteurs propres. La matrice de transformation P est composée des vecteurs propres comme colonnes:
P = [[1/√2 -1/√2]
     [1/√2  1/√2]]
### En multipliant les données par la matrice de transformation, on obtient les données transformées:
X_transformed = P * X
## 3. Composantes principales:
### Les composantes principales sont les nouvelles variables créées par la transformation PCA. Elles sont une combinaison linéaire des variables originales.
## 4. Variance expliquée:
### La variance expliquée par chaque composante principale est une mesure de son importance. Elle est calculée comme le rapport entre la valeur propre de la composante et la somme de toutes les valeurs propres.
# Exemple concret
### Supposons que nous avons un ensemble de données composé de deux variables: la taille et le poids d'un groupe de personnes. Nous pouvons utiliser le PCA pour réduire la dimensionnalité de ces données en les projetant sur une seule dimension.
## 1. Calculer la matrice de covariance:

```python
import numpy as np

# Définition de la matrice C (matrice de covariance)
C = np.array([[1, 0.5],
              [0.5, 1]])
```

## 2. Calculer les vecteurs propres et les valeurs propres:

```python
from math import sqrt

# Définition des vecteurs propres
v1 = [1/sqrt(2), 1/sqrt(2)]
v2 = [-1/sqrt(2), 1/sqrt(2)]

# Définition des valeurs propres
λ1 = 2
λ2 = 0


P = np.array([v1, v2]) # Construction de la matrice P contenant les vecteurs propres

X = np.random.rand(2, 2)

X
```

## 3. Transformer les données:

```python
X_transformed = P * X
X_transformed
```

## 5. Visualiser les résultats:

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Tracer la matrice X
axs[0].scatter(X[0], X[1])
axs[0].set_title('Matrice X')
axs[0].set_xlabel('X1')
axs[0].set_ylabel('X2')
axs[0].grid(True)
axs[0].axis('equal')

# Tracer la matrice X_transformed
axs[1].scatter(X_transformed[0], X_transformed[1], color='r')
axs[1].set_title('Matrice X_transformed')
axs[1].set_xlabel('Component 1')
axs[1].set_ylabel('Component 2')
axs[1].grid(True)
axs[1].axis('equal')

plt.tight_layout()
plt.show()
```

# Appliquons cette technique sur notre dataset diabetes

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Charger les données
data=pd.read_csv('diabetes.csv')
data=data.head(40)
data
```

# Séparer les caractéristiques (features) et les étiquettes (labels)

```python

X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

# Standardisation des données

```python

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Réduction de dimension avec PCA

```python

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

# Création d'un DataFrame pour les données non réduites

```python

original_df = X.copy()
original_df['Outcome'] = y
```

# Création d'un DataFrame pour les données réduites

```python

pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Outcome'] = y
```

# Visualisation des données avant et après PCA

```python

fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Données après PCA
axs[0].scatter(pca_df.loc[pca_df['Outcome'] == 0, 'PC1'], pca_df.loc[pca_df['Outcome'] == 0, 'PC2'], color='blue', label='Non-diabetic')
axs[0].scatter(pca_df.loc[pca_df['Outcome'] == 1, 'PC1'], pca_df.loc[pca_df['Outcome'] == 1, 'PC2'], color='red', label='Diabetic')
axs[0].set_xlabel('Principal Component 1')
axs[0].set_ylabel('Principal Component 2')
axs[0].set_title('PCA - Projection des données sur les deux premiers composants principaux')
axs[0].legend()

# Données originales
axs[1].scatter(original_df.loc[original_df['Outcome'] == 0, 'Glucose'], original_df.loc[original_df['Outcome'] == 0, 'BMI'], color='blue', label='Non-diabetic')
axs[1].scatter(original_df.loc[original_df['Outcome'] == 1, 'Glucose'], original_df.loc[original_df['Outcome'] == 1, 'BMI'], color='red', label='Diabetic')
axs[1].set_xlabel('Glucose')
axs[1].set_ylabel('BMI')
axs[1].set_title('Données Originales - Glucose vs BMI')
axs[1].legend()

plt.tight_layout()
plt.show()
```
